```python
def new_game():
    gusses = []
    correct_guesses = 0
    question_num = 1
    
    for key in questions:
        print(key)
        for i in options[question_num - 1]:
            print(i)
        guess = input("Enter A, B, C or D : ")
        guess = guess.upper()
        gusses.append(guess)
        
        correct_guesses = correct_guesses + check_answer(questions.get(key),guess)
        question_num = question_num + 1
        print("--------------------------")
        
    
    display_score(correct_guesses, gusses)
    
#---------------------------
def check_answer(answer,guess):
    
    if answer == guess:
        print("CORRECT ANSWER !!!")
        return 1  # 1 point for right answer
    else:
        print("Wrong Answer Buddy :( ")
        return 0 # no points = 0 for wrong answers

#---------------------------
def display_score(correct_gusses, guesses):
    print("--------------------------")
    print("RESULTS")
    print("--------------------------")
    
    print("Correct Answers for the above 4 questions :", end="")
    for i in questions:
        print(questions.get(i), end=" ")
    print()
    
    print("Guesses you have made for the above 4 questions :", end="")
    for i in guesses:
        print(i, end=" ")
    print()    
    
    score = ((correct_gusses/len(questions)) * 100 )
    print("Yor total score is :",str(score),"%")
    
#---------------------------
def play_again():
    response = input("Do you want to play again? (Yes/No) :")
    response = response.upper()
    
    if response == "YES":
        return True
    else:
        return False
#---------------------------


# Dictionary object for Questions and Answers, since dict is an key value pair
questions = {
    "How many teams in NBA?:":"B",
    "Which Year NBA was first started?:":"A",
    "How many players are there in a basketball team?:":"C",
    "What is the standard for a basketball in the NBA (in inches)?:":"D"
}

# Collections to Hold all the different possible answers for a MCQ - A 2-D List will be Perfect for this
options=[["A. 29 ","B. 30","C. 31","D. 32"],
         ["A. 1946","B. 1965","C. 1947","D. 1948"],
         ["A. 10","B. 11","C. 12","D. 13"],
         ["A. 30","B. 29","C. 28.5","D. 29.5"]]


new_game()

while play_again():
    new_game()
    
print("Byeeeeee!!!")
```
